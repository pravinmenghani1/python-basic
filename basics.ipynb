#Python Optimization

Reusing frequently used immutable objects:

Saves memory
Improves performance
Avoids creating duplicate objects unnecessarily

Since integers are immutable, sharing them is completely safe.

$Modulo

```python
name = "Pravin"
score = 98.567
subjects = int(input("Enter the number of subjects: "))

print("Student: %s, Score: %f, Subject: %a" %(name, score, subjects))

print("Student: %s, Score: %.2f, Subject: %a" %(name, score, subjects))

print("Student: %s, Score: %8.2f, Subject: %a" %(name, score, subjects))

print("Student: %s, Score: %05d, Subject: %a" %(name, score, subjects))

```
