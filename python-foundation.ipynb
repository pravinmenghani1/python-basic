## 5. The Print Statement

`print()` is Python's way of displaying output.

**Key parameters:** `sep=" "` (separator), `end="\n"` (line ending), `file=sys.stdout`, `flush=False`

````python
# Basic print
print("Hello, World!")

# Print multiple values (comma-separated)
print("Name:", "Pravin", "Age:", 30)

# Custom separator
print("Python", "is", "awesome", sep=" 🐍 ")

# Custom end (default is newline \n)
print("Loading", end="...")
print("Done!")

# Print numbers and expressions
print(2 + 3)       # Evaluates expression first
print(10 / 3)      # Float division
print(10 // 3)     # Integer (floor) division

# Print multiple types
name = "Alice"
age  = 25
gpa  = 3.87
print(name, age, gpa)
````

## 6. User Input & Square Root Calculator

> ⚠️ `input()` **always returns a string**. You must convert it with `int()` or `float()` before doing math. Forgetting this is the #1 beginner bug!

````python
import math

# input() always returns a STRING — we must convert
number_str = input("Enter a number: ")   # e.g., user types "25"
number = float(number_str)

if number < 0:
    print("❌ Cannot compute square root of a negative number.")
else:
    result = math.sqrt(number)
    print(f"✅ Square root of {number} is {result:.4f}")

# Compact one-liner version
num = float(input("Enter a number: "))
print(f"√{num} = {num ** 0.5:.4f}")   # ** 0.5 is also square root
````

## 8. Indentation

> ⚠️ **Python's Golden Rule:** Unlike most languages that use `{ }` braces, Python uses **indentation** (whitespace) to define code blocks. Wrong indentation = `IndentationError`.

- Use **4 spaces** per level (PEP 8 standard)
- Never mix spaces and tabs
- All blocks use indentation: `if`, `for`, `while`, `def`, `class`, `with`, `try`

````python
# ✅ CORRECT: 4 spaces per level (PEP 8 standard)
def greet(name):
    if name:                        # Level 1 — 4 spaces
        message = "Hello, " + name  # Level 2 — 8 spaces
        print(message)
    else:
        print("No name provided")

greet("Pravin")

# Nested loops — each level adds 4 spaces
for i in range(3):
    for j in range(3):
        if i == j:
            print(f"Diagonal: ({i},{j})")

# All blocks use indentation
try:
    result = 10 / 0
except ZeroDivisionError:
    print("Cannot divide by zero!")
````

## 9. Identifiers

An **identifier** is a name you give to variables, functions, classes, or modules.

**Rules:**
- Can contain letters (a–z, A–Z), digits (0–9), and underscore `_`
- Must start with a letter or underscore — **never a digit**
- Case-sensitive: `score`, `Score`, `SCORE` are three different identifiers
- Cannot be a Python keyword (`if`, `for`, `class`…)

**Conventions (PEP 8):** variables & functions → `snake_case` | classes → `PascalCase` | constants → `ALL_CAPS`

````python
# ✅ Valid identifiers
student_name = "Riya"
_private_var = 42
totalMarks   = 95
MAX_SIZE     = 1000

# ❌ Invalid identifiers (would cause SyntaxError)
# 2fast = "car"     # starts with digit
# my-var = 10       # hyphen not allowed
# for = 5           # reserved keyword

# Case sensitivity demonstration
color = "red"
Color = "blue"
COLOR = "green"
print(color, Color, COLOR)   # Three distinct variables!

import keyword
print(keyword.iskeyword("for"))    # True
print(keyword.iskeyword("total"))  # False
````

## 11. Variables in Python

> **Analogy:** A variable is like a labelled sticky note on a memory slot. Python is **dynamically typed** — the same variable can hold any type at any time.

````python
# Basic assignment
name  = "Pravin"
age   = 28
score = 98.5

# Dynamic typing — same variable, different types
x = 10;        print(type(x))  # <class 'int'>
x = "hello";   print(type(x))  # <class 'str'>
x = [1, 2, 3]; print(type(x))  # <class 'list'>

# Multiple assignment (tuple unpacking)
a, b, c = 1, 2, 3
print(a, b, c)

# Same value to multiple variables
p = q = r = 0
print(p, q, r)

# Swap without temp variable
x, y = 10, 20
x, y = y, x
print(f"After swap: x={x}, y={y}")

# Augmented assignment
counter = 0
counter += 1
counter *= 3
print(counter)  # 3

# Delete a variable
temp = "temporary"
del temp
# print(temp)  # ← NameError
````

## 12. String Interpolation

Python gives you **4 ways** to embed variables inside strings:

| Method | Syntax | Best For | Version |
|---|---|---|---|
| `%` modulo | `"Hi %s" % name` | Legacy code | All |
| `str.format()` | `"Hi {}".format(name)` | Complex alignment | 2.7+ |
| **f-strings** | `f"Hi {name}"` | **Everything — use by default** | 3.6+ |
| `Template` | `Template("Hi $name")` | User-facing templates (safe) | All |

````python
# Method 1 — % (Modulo / Old-style formatting)
name  = "Pravin"
score = 98.567

print("Student: %s, Score: %d" % (name, score))  # %s=str, %d=int
print("Score: %.2f" % score)      # 2 decimal places
print("Score: %8.2f" % score)     # right-align in width 8
print("Roll No: %05d" % 42)       # zero-padded → 00042
````

````python
# Method 2 — str.format()
name  = "Pravin"
score = 98.567

print("Hello, {}! Your score is {:.2f}".format(name, score))
print("{0} said: '{1} is {0}'".format("Python", "Amazing"))
print("{student} scored {marks:.1f}".format(student="Riya", marks=95.5))
print("{:<10} {:>10} {:^10}".format("left", "right", "center"))
print("{:,}".format(1000000))    # thousand separator
print("{:.2%}".format(0.876))   # percentage
````

````python
# Method 3 — f-strings (Python 3.6+) ← RECOMMENDED
name  = "Pravin"
score = 98.567
items = ["Python", "AI", "ML"]

print(f"Hello, {name}! Score: {score:.2f}")
print(f"Double score: {score * 2:.1f}")       # expressions inside {}
print(f"Upper name: {name.upper()}")
print(f"Result: {'Pass ✅' if score > 40 else 'Fail ❌'}")

x = 42
print(f"{x = }")   # Python 3.8+ self-documenting

print(f"Languages: {', '.join(items)}")
````

````python
# Method 4 — Template class (string module)
from string import Template

t = Template("Hello, $name! You scored $score in $subject.")
print(t.substitute(name="Riya", score=95, subject="Python"))

# safe_substitute — doesn't crash on missing keys
t2 = Template("Dear $name, your order $order_id is ready.")
print(t2.safe_substitute(name="Alice"))  # order_id missing — no error
````

## 13. Standard Data Types & Mutability

| Type | Category | Mutability | Example |
|---|---|---|---|
| `int` | Numeric | Immutable | `x = 42` |
| `float` | Numeric | Immutable | `pi = 3.14` |
| `complex` | Numeric | Immutable | `z = 3+4j` |
| `str` | Text | Immutable | `s = "hello"` |
| `bool` | Boolean | Immutable | `flag = True` |
| `tuple` | Sequence | Immutable | `t = (1,2,3)` |
| `list` | Sequence | **Mutable** | `l = [1,2,3]` |
| `dict` | Mapping | **Mutable** | `d = {"a":1}` |
| `set` | Set | **Mutable** | `s = {1,2,3}` |
| `frozenset` | Set | Immutable | `fs = frozenset({1,2})` |
| `NoneType` | Null | Immutable | `x = None` |

> **Immutable** = once created, value cannot change in-place (creates a new object).  
> **Mutable** = you can modify the object in-place.  
> *Analogy:* Immutable = printed book (reprint to change). Mutable = whiteboard (erase & rewrite).

````python
# Demonstrating mutability

# Immutable — str: modifying creates a NEW object
s = "hello"
print(id(s))
s = s + " world"   # new string object created
print(id(s))       # different id!

# Mutable — list: modified in-place, same object
lst = [1, 2, 3]
print(id(lst))
lst.append(4)      # in-place modification
print(id(lst))     # same id!
print(lst)

# Check type of any value
values = [42, 3.14, "hello", True, None, [1,2], (1,2), {"a":1}, {1,2}]
for v in values:
    print(f"{repr(v):20} → {type(v).__name__}")
````

## 14. Numbers — Types, Instances & Type Conversion

````python
import sys

# ---- Integer (int) ----
a = 42
print(type(a))             # <class 'int'>
print(isinstance(a, int))  # True — isinstance() is preferred over type()==
print(sys.getsizeof(a))    # memory size in bytes

# Python int has UNLIMITED precision (unlike C/Java)
big = 10 ** 100            # Googol — no overflow!
print(big)

# ---- Float ----
b = 3.14
print(type(b))             # <class 'float'>
print(b.is_integer())      # False

# Float precision gotcha
print(0.1 + 0.2)           # 0.30000000000000004 ← floating point error!
from decimal import Decimal
print(Decimal("0.1") + Decimal("0.2"))  # 0.3 (exact)

# ---- Complex ----
z = 3 + 4j
print(type(z))             # <class 'complex'>
print(z.real, z.imag)      # 3.0  4.0
print(abs(z))              # magnitude = √(3²+4²) = 5.0

# ---- Type Conversion ----
print(int("42"))           # str → int
print(float("3.14"))       # str → float
print(str(100))            # int → str
print(bool(0), bool(1))    # int → bool

print(int("FF", 16))       # hex string → int: 255
print(int("1010", 2))      # binary string → int: 10
print(bin(255))            # int → binary string
print(hex(255))            # int → hex string
````

## 15. Collections — Strings, Tuples, Lists, Dicts, Sets

| Type | Ordered? | Unique? | Mutable? | Indexed? | Best For |
|---|---|---|---|---|---|
| `str` | ✅ | ❌ | ❌ | ✅ | Text, messages |
| `tuple` | ✅ | ❌ | ❌ | ✅ | Fixed records, GPS coords |
| `list` | ✅ | ❌ | ✅ | ✅ | Dynamic collections, stacks |
| `dict` | ✅(3.7+) | Keys only | ✅ | By key | Lookup tables, JSON, config |
| `set` | ❌ | ✅ | ✅ | ❌ | Deduplication, membership tests |

````python
# 🔤 STRING — Immutable sequence of characters
s = "Hello, Python World!"

# Indexing & Slicing
print(s[0])        # H  (first char)
print(s[-1])       # !  (last char)
print(s[0:5])      # Hello
print(s[::2])      # Every other character
print(s[::-1])     # Reverse the string

# String methods
print(s.lower())
print(s.upper())
print(s.replace("Python", "AI"))
print(s.split(", "))          # Split into list
print("  strip me  ".strip())
print(s.count("l"))           # Count occurrences
print(s.startswith("Hello"))  # True/False
print(s.find("Python"))       # Index of first occurrence

# Real-world: parsing a CSV line
csv_line = "Alice,28,Engineer,Pune"
parts = csv_line.split(",")
print(f"Name: {parts[0]}, Age: {parts[1]}, Role: {parts[2]}")
````

````python
# 📦 TUPLE — Immutable ordered sequence
point  = (3, 4)
colors = ("red", "green", "blue")
single = (42,)   # Single-element tuple needs trailing comma!

# Tuple unpacking
x, y = point
lat, lon = 18.5204, 73.8567  # GPS coordinates
print(f"Pune: lat={lat}, lon={lon}")

# Tuples as dictionary keys (lists cannot be dict keys)
locations = {
    (18.5204, 73.8567): "Pune",
    (28.6139, 77.2090): "Delhi"
}
print(locations[(18.5204, 73.8567)])

# Named tuple — self-documenting
from collections import namedtuple
Student = namedtuple("Student", ["name", "age", "gpa"])
s = Student("Pravin", 28, 3.9)
print(s.name, s.gpa)

# Function returning multiple values (returns a tuple)
def min_max(data):
    return min(data), max(data)

lo, hi = min_max([5, 2, 8, 1, 9])
print(f"Range: {lo} – {hi}")
````

````python
# 📋 LIST — Mutable ordered sequence
fruits = ["apple", "banana", "cherry"]

fruits.append("date")          # Add to end
fruits.insert(1, "avocado")    # Insert at index
fruits.remove("banana")        # Remove by value
popped = fruits.pop()          # Remove & return last
print(fruits)

# Sorting
numbers = [5, 2, 8, 1, 9, 3]
numbers.sort()                 # In-place sort
print(numbers)
print(sorted([5, 2, 8], reverse=True))  # Returns new sorted list

# List comprehension
squares = [x**2 for x in range(1, 6)]
print(squares)

evens = [x for x in range(20) if x % 2 == 0]
print(evens)

# Nested list (2D matrix)
matrix = [[1,2,3],[4,5,6],[7,8,9]]
print(matrix[1][2])   # Row 1, Col 2 → 6

# Real-world: student scores
scores = [72, 85, 91, 63, 78, 95, 88]
passed = [s for s in scores if s >= 75]
avg    = sum(scores) / len(scores)
print(f"Passed: {len(passed)}/{len(scores)}, Avg: {avg:.1f}")
````

````python
# 🗂️ DICTIONARY — Mutable key-value mapping
student = {
    "name"   : "Pravin",
    "age"    : 28,
    "courses": ["AI", "Python", "ML"],
    "gpa"    : 3.9
}

print(student["name"])
print(student.get("phone", "N/A"))   # Safe access with default

student["email"] = "pravin@ai.com"   # Add new key
student["age"]   = 29                # Update existing
del student["gpa"]                   # Delete key

for key, value in student.items():
    print(f"  {key}: {value}")

# Dictionary comprehension
word_lengths = {word: len(word) for word in ["Python", "AI", "Machine"]}
print(word_lengths)

# Counting with dict
votes = ["Alice","Bob","Alice","Charlie","Bob","Alice"]
count = {}
for v in votes:
    count[v] = count.get(v, 0) + 1
print(count)
````

````python
# 🔵 SET — Mutable unordered collection of unique elements

# Set automatically removes duplicates
tags = {"python", "ai", "python", "ml", "ai", "python"}
print(tags)   # Only unique values

# Set operations (mathematical)
A = {1, 2, 3, 4, 5}
B = {3, 4, 5, 6, 7}

print("Union:       ", A | B)   # All elements
print("Intersection:", A & B)   # Common elements
print("Difference:  ", A - B)   # In A, not in B
print("Sym. Diff:   ", A ^ B)   # In A or B, not both

# Membership test (O(1) — much faster than list!)
primes = {2, 3, 5, 7, 11, 13}
print(7 in primes)   # True
print(6 in primes)   # False

# Real-world 1: Remove duplicates from a list
raw_ids = [101, 203, 101, 305, 203, 101, 400]
unique_ids = list(set(raw_ids))
print("Unique IDs:", unique_ids)

# Real-world 2: Common friends
alice_friends = {"Bob", "Charlie", "Diana", "Eve"}
bob_friends   = {"Charlie", "Frank", "Diana", "Grace"}
mutual = alice_friends & bob_friends
print("Mutual friends:", mutual)

# Frozenset — immutable set, usable as dict key
fs = frozenset({1, 2, 3})
d  = {fs: "immutable set as key"}
print(d)
````

## 16. Operators & Precedence

**Operator Precedence** (highest → lowest):

| Priority | Operator(s) | Description |
|---|---|---|
| HIGHEST | `()` | Parentheses |
| | `**` | Exponentiation (right-associative) |
| | `+x -x ~x` | Unary operators |
| | `* / // %` | Mul, Div, Floor Div, Modulo |
| | `+ -` | Addition, Subtraction |
| | `<< >>` | Bitwise shift |
| | `&` | Bitwise AND |
| | `^` | Bitwise XOR |
| | `\|` | Bitwise OR |
| | `== != < > <= >= is is not in not in` | Comparisons |
| | `not` | Logical NOT |
| | `and` | Logical AND |
| LOWEST | `or` | Logical OR |

````python
# 1. ARITHMETIC OPERATORS
a, b = 15, 4

print(a + b)    # Addition       → 19
print(a - b)    # Subtraction    → 11
print(a * b)    # Multiplication → 60
print(a / b)    # True division  → 3.75 (always float)
print(a // b)   # Floor division → 3
print(a % b)    # Modulo         → 3  (15 = 4×3 + 3)
print(a ** b)   # Exponentiation → 50625
````

````python
# 2. COMPARISON / RELATIONAL OPERATORS
a, b = 15, 4

print(a == b)   # Equal          → False
print(a != b)   # Not equal      → True
print(a >  b)   # Greater        → True
print(a <  b)   # Less           → False
print(a >= b)   # Greater/equal  → True
print(a <= b)   # Less/equal     → False

# Chained comparisons (Python exclusive!)
age = 25
print(18 <= age <= 65)       # True — voter age check
print(1 < 2 < 3 < 4 < 5)    # True
````

````python
# 3. LOGICAL OPERATORS
x, y = True, False

print(x and y)   # Both must be True → False
print(x or  y)   # At least one True → True
print(not  x)    # Invert           → False

# Short-circuit evaluation
print(False and 1/0)   # 1/0 never evaluated! → False
print(True  or  1/0)   # 1/0 never evaluated! → True
````

````python
# 4. BITWISE OPERATORS
p, q = 0b1010, 0b1100    # 10, 12 in decimal

print(bin(p & q))    # AND         → 0b1000 (8)
print(bin(p | q))    # OR          → 0b1110 (14)
print(bin(p ^ q))    # XOR         → 0b0110 (6)
print(bin(~p))       # NOT         → complement
print(bin(p << 1))   # Left shift  → 0b10100 (20)
print(bin(p >> 1))   # Right shift → 0b0101 (5)
````

````python
# 5. ASSIGNMENT OPERATORS
n  = 10
n += 5;  print(n)   # 15
n -= 3;  print(n)   # 12
n *= 2;  print(n)   # 24
n //= 5; print(n)   # 4
n **= 3; print(n)   # 64
n %= 10; print(n)   # 4
````

````python
# 6. IDENTITY OPERATORS
a = [1, 2, 3]
b = [1, 2, 3]
c = a

print(a == b)      # True  — same VALUES
print(a is b)      # False — different OBJECTS in memory
print(a is c)      # True  — c points to same object as a
print(a is not b)  # True

# 7. MEMBERSHIP OPERATORS
fruits = ["apple", "banana", "cherry"]
print("apple"  in fruits)     # True
print("mango" not in fruits)  # True
print("e" in "hello")         # True (works on strings)
print(3 in {1, 2, 3, 4})      # True (works on sets)
````

````python
# OPERATOR PRECEDENCE DEMOS

print(2 + 3 * 4)       # 14 (not 20!) — * before +
print((2 + 3) * 4)     # 20 — parentheses override

print(2 ** 3 ** 2)     # 512 — ** is RIGHT-associative: 2**(3**2)
print((2 ** 3) ** 2)   # 64  — left-grouping

# Chained comparisons
age = 25
print(18 <= age <= 65)   # True — Pythonic range check

# Ternary operator
a = 0
result = "non-zero" if a else "zero"
print(result)

# Modulo use cases
print(15 % 3 == 0)   # True  — divisibility check
print(17 % 2)        # 1     — odd number
hour = 14 % 12       # 12-hour format
print(f"{hour} PM")
````
