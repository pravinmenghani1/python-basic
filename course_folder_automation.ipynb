# Course Folder Automation
**Use Case:** John films and edits online courses. He needs an automated way to create organized folder structures for each course, lesson, and file type — instead of doing it manually.

## Step 1: Import Required Module
Python's built-in `os` module lets us create folders and navigate the file system.

```python
import os
```

## Step 2: Define Course Configuration
John specifies the courses, number of lessons per course, and the file type folders he needs.

```python
# Courses and their lesson counts
courses = {
    "Physics": 5,
    "Chemistry": 4,
    "Mathematics": 6
}

# File type subfolders inside each lesson
file_types = ["Video", "Audio", "Motion Graphics"]

# Root directory where all course folders will be created
root = "John_Courses"
```

## Step 3: Automate Folder Creation
Loop through each course and lesson, then create the required subfolders.

```python
for course, num_lessons in courses.items():
    for lesson_num in range(1, num_lessons + 1):
        lesson_folder = f"Lesson_{lesson_num}_of_{num_lessons}"
        for file_type in file_types:
            path = os.path.join(root, course, lesson_folder, file_type)
            os.makedirs(path, exist_ok=True)

print("✅ All course folders created successfully!")
```

## Step 4: Verify the Folder Structure
Walk through the created directories and print the structure.

```python
for dirpath, dirnames, filenames in os.walk(root):
    depth = dirpath.replace(root, "").count(os.sep)
    indent = "    " * depth
    print(f"{indent}{os.path.basename(dirpath)}/")
```

## Step 5: Reusable Function
Wrap everything into a function so John can reuse it anytime with different inputs.

```python
def create_course_structure(courses: dict, file_types: list, root: str = "Courses"):
    """
    courses   : {"CourseName": num_lessons}
    file_types: list of subfolder names per lesson
    root      : top-level directory name
    """
    for course, num_lessons in courses.items():
        for lesson_num in range(1, num_lessons + 1):
            lesson_folder = f"Lesson_{lesson_num}_of_{num_lessons}"
            for file_type in file_types:
                path = os.path.join(root, course, lesson_folder, file_type)
                os.makedirs(path, exist_ok=True)
    print(f"✅ Structure created under '{root}/'")


# Example: add a new course anytime
create_course_structure(
    courses={"Biology": 3, "History": 4},
    file_types=["Video", "Audio", "Motion Graphics"],
    root="John_Courses"
)
```

## Summary

| Concept Used | Purpose |
|---|---|
| `os.makedirs` | Creates nested folders in one call |
| `exist_ok=True` | Skips error if folder already exists |
| `os.path.join` | Builds cross-platform file paths |
| `os.walk` | Traverses and prints folder tree |
| Dictionary | Stores course → lesson count mapping |
| f-strings | Dynamically names lesson folders |
| Function | Makes the solution reusable |
