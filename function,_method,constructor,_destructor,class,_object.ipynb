```python
def welcome_patient(patient_name: str) -> tuple:
    """
    Standard Procedure: Patient Welcome
    Called for EVERY new patient arrival.
    Returns a tuple with the message and the patient's name.
    """
    message = ("Welcome to City General Hospital", patient_name)
    return message

# --- Hardcoded call (testing the SOP in isolation) ---
print(welcome_patient("Rajesh Sharma"))
# Output: ('Welcome to City General Hospital', 'Rajesh Sharma')

# --- Dynamic call (patient name from reception form) ---
name = input("Enter patient name: ")
result = welcome_patient(name)
print(result)
# Output: ('Welcome to City General Hospital', 'Priya Nair')
```

```python
def calculate_bmi(weight_kg: float, height_m: float) -> float:
    """Standard BMI formula — used by every department."""
    return weight_kg / (height_m ** 2)

def is_fever(temperature_celsius: float) -> bool:
    """Returns True if patient has fever (>= 37.5°C)."""
    return temperature_celsius >= 37.5

def format_patient_id(ward: str, bed_number: int) -> str:
    """Generates a standard patient ID: WARD-BEDNUMBER."""
    return f"{ward.upper()}-{bed_number:03d}"

print(calculate_bmi(70, 1.75))         # 22.86
print(is_fever(38.2))                  # True
print(format_patient_id("icu", 7))    # ICU-007
```

```python
# patient_blueprint.py — Class = Blank Patient Record Form

class Patient:
    """
    Blueprint for every patient record in City General Hospital.
    This is NOT a patient. This is the DESIGN of what every
    patient record must look like.
    """

    # --- Action 1: Admit the patient ---
    def admit(self):
        print(f"Patient {self.name} admitted to {self.ward} ward.")

    # --- Action 2: Record a diagnosis ---
    def diagnose(self):
        print(f"Patient {self.name} diagnosed: {self.diagnosis}")

    # --- Action 3: Discharge the patient ---
    def discharge(self):
        print(f"Patient {self.name} discharged. Bill: ₹{self.bill:,}")


class Patient:
    def admit(self):    print(f"[ADMITTED] {self.name} → {self.ward} Ward")
    def diagnose(self): print(f"[DIAGNOSIS] {self.name}: {self.diagnosis}")
    def discharge(self): print(f"[DISCHARGED] {self.name} | Bill: ₹{self.bill:,}")

# ── Receptionist fills the form for Patient 1 ──
rajesh = Patient()            # Pick up a blank form
rajesh.patient_id = "CGH-001"
rajesh.name       = "Rajesh Sharma"
rajesh.age        = 45
rajesh.blood_group= "B+"
rajesh.diagnosis  = "Chest Pain — Cardiac Evaluation"
rajesh.ward       = "Cardiology"
rajesh.bill       = 28500

# ── Receptionist fills the form for Patient 2 ──
priya = Patient()             # Another blank form — fresh copy
priya.name = "Priya Nair"; priya.ward = "General"; priya.bill = 4200

rajesh.admit()     # [ADMITTED] Rajesh Sharma → Cardiology Ward
priya.admit()      # [ADMITTED] Priya Nair → General Ward
rajesh.diagnose()  # [DIAGNOSIS] Rajesh Sharma: Chest Pain — Cardiac Evaluation
priya.discharge()  # [DISCHARGED] Priya Nair | Bill: ₹4,200

```

```python
class Patient:
    """Every patient MUST be registered with all mandatory details."""

    def __init__(self, patient_id: str, name: str, age: int,
                 blood_group: str, ward: str):
        # Runs AUTOMATICALLY the moment a patient is admitted
        self.patient_id  = patient_id   # Mandatory: unique ID
        self.name        = name          # Mandatory: full name
        self.age         = age           # Mandatory: age
        self.blood_group = blood_group   # Mandatory: critical for surgery
        self.ward        = ward          # Mandatory: which ward
        self.diagnosis   = None         # Set later by doctor
        self.bill        = 0.0          # Accumulates as services used
        print(f"[ADMISSION] ✅ {self.name} → {self.ward} | ID: {self.patient_id}")

    def diagnose(self, diagnosis: str):
        self.diagnosis = diagnosis
        print(f"[DOCTOR] {self.name} diagnosed: {self.diagnosis}")

    def add_charge(self, service: str, amount: float):
        self.bill += amount
        print(f"[BILLING] {service}: ₹{amount:,} | Total: ₹{self.bill:,}")

    def __del__(self):
        print(f"[ARCHIVE] Record for {self.name} archived.")

# Patient walks in → __init__ fires automatically
rajesh = Patient("CGH-001", "Rajesh Sharma", 45, "B+", "Cardiology")
# [ADMISSION] ✅ Rajesh Sharma → Cardiology | ID: CGH-001

rajesh.diagnose("Acute Myocardial Infarction — Requires Stenting")
rajesh.add_charge("Emergency Consultation", 2500)
rajesh.add_charge("Angiography", 18000)
rajesh.add_charge("Stenting Procedure", 45000)
# [BILLING] Stenting Procedure: ₹45,000 | Total: ₹65,500
del rajesh
# [ARCHIVE] Record for Rajesh Sharma archived.
```

#Methods

```python
class Patient:
    def __init__(self, patient_id, name, age, blood_group, ward, daily_charges):
        self.patient_id    = patient_id
        self.name          = name
        self.age           = age
        self.ward          = ward
        self.daily_charges = daily_charges   # List of per-day charges
        self.diagnosis     = None
        self.medications   = []

    # Method 1: Display patient summary (Nurse station view)
    def display(self):
        print(f"  {self.patient_id} | {self.name} | {self.ward} | {self.diagnosis}")

    # Method 2: Compute average daily charge (Finance dept)
    def average_daily_charge(self) -> float:
        return sum(self.daily_charges) / len(self.daily_charges)

    # Method 3: Triage priority (calls another method!)
    def triage_priority(self):
        avg = self.average_daily_charge()    # Calls another method
        if   avg >= 10000: print(f"  {self.name} → 🔴 CRITICAL")
        elif avg >= 5000:  print(f"  {self.name} → 🟡 HIGH")
        else:              print(f"  {self.name} → 🟢 STANDARD")

    # Method 4: Add medication to chart
    def prescribe(self, medicine: str):
        self.medications.append(medicine)
        print(f"  [Rx] {medicine} prescribed to {self.name}")

    def __del__(self): print(f"  [ARCHIVE] {self.name}'s record closed.")


arjun = Patient("CGH-003", "Arjun Mehta", 58, "A+", "Neurology",
                [8500, 9200, 11000, 7800, 12500, 10300])
arjun.diagnosis = "Ischemic Stroke — Post-op Recovery"

arjun.display()
print(f"  Average: ₹{arjun.average_daily_charge():,.0f}/day")
# Average: ₹9,883/day
arjun.triage_priority()
# Arjun Mehta → 🔴 CRITICAL
arjun.prescribe("Aspirin 75mg once daily")
```

**Function**
A function is a reusable block of code defined independently using def.

**Method**

A method is a function that is defined inside a class and is associated with that class or its objects.

In short: Every method is a function, but not every function is a method.

```python
def print_school_name(): #function defined outside of the class
  print("School Name: ABC Public School")

class student:
  def __init__(self,name,rollno,marks):
    self.name=name
    self.rollno=rollno
    self.marks=marks

  def average(self):
    return sum(self.marks)/len(self.marks)

  def display(self):
    print("name:",self.name)
    print("rollno:",self.rollno)
    print("marks:",self.marks)
    print

  def grade(self):
    if self.average()>=90:
      print("grade A")

    elif self.average()>=80:
      print("grade B")

    elif self.average()>=70:
      print("grade C")

    elif self.average()>=60:
      print("grade D")

    else:
      print("grade E")

  def __del__(self):
    print("object destroyed")

first_student = student("varun",1,[50,80,60,70,80,90])
first_student.display()
first_student.grade()
del first_student

print_school_name()
```

#Calling a method from another class

```python
class Calculator:
    def add(self, a, b):
        return a + b


class Report:
    def show_result(self):
        calc = Calculator()          # Create object of first class
        result = calc.add(10, 20)    # Call method from Calculator
        print("Result:", result)


report = Report()
report.show_result()
```

Why Do We Use __del__()? What Is Its Purpose?

__del__() is called the destructor method.

It is executed when an object is about to be destroyed (for example, when you use del obj or when Python removes the object from memory).

```python
class Student:
    def __init__(self, name):
        self.name = name
        print("Object created")

    def __del__(self):
        print(f"Object for {self.name} destroyed")


s = Student("Varun")
del s
```
