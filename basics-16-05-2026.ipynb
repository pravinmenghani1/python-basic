## 13. Standard Data Types & Mutability

| Type | Category | Mutability | Example |
|---|---|---|---|
| `int` | Numeric | Immutable | `x = 42` |
| `float` | Numeric | Immutable | `pi = 3.14` |
| `complex` | Numeric | Immutable | `z = 3+4j` |
| `str` | Text | Immutable | `s = "hello"` |
| `bool` | Boolean | Immutable | `flag = True` |
| `tuple` | Sequence | Immutable | `t = (1,2,3)` |
| `list` | Sequence | **Mutable** | `l = [1,2,3]` |
| `dict` | Mapping | **Mutable** | `d = {"a":1}` |
| `set` | Set | **Mutable** | `s = {1,2,3}` |
| `frozenset` | Set | Immutable | `fs = frozenset({1,2})` |
| `NoneType` | Null | Immutable | `x = None` |

> **Immutable** = once created, value cannot change in-place (creates a new object).  
> **Mutable** = you can modify the object in-place.  
> *Analogy:* Immutable = printed book (reprint to change). Mutable = whiteboard (erase & rewrite).

#SETS

**A set stores only unique items and automatically removes duplicates**

```python
numbers = {1,2,3,4,5,6,2,3,4}

print(numbers)

numbers.add(7)

print(numbers)

numbers.add(2) #add another has no impact

print("The set numbers is : ", numbers)

numbers[4] = 8

'''
A set is an unordered collection of unique elements

1. Duplicates are automatically removed
2. Order is not preserved
3. Indexing is not supported

'''

print(numbers)
```

```python
# SET EXAMPLE WITH STRINGS

cities = {"Pune", "Mumbai", "Delhi", "Chennai", "Pune"}

print(cities)

cities.add("Bangalore")
print(cities)

cities.remove("Mumbai")
print(cities)

cities.append("Hyderabad") #wont work and error
print(cities)
```

Thin of set as a bag of unique marbles inside it.
1. You can add marbles
2. You can remobe marbles
3. You cannot say "replace the marble at position 4" because its unordered list

#LOOPS

```python
#While loop: it will repeat as long as condition is true

count = 1
while count <= 5:
  print(f'Count: {count}')
  count += 1
```

```python
#For loop

fruits = ['apple', 'grapes', 'banana']
for fruit in fruits:
  print(fruit)

# using range

for i in range(1,5):
  print(i)
```

```python
'''FOR-ELSE CONCEPT
else runs - no break hit
'''

for n in range(3):
  print(n)
else:
  print('Loop finished successfully')


# else doesnt run - break hit

for n in range(3):
  print(n)
  if n == 2:
    break
else:
  print('Loop finished successfully and this may not print')
```

#Real world use case of for--else statement: in searching

```python
num = [10,20,30,40]
target = 25

for n in num:
  if num == target:
    print("Found")
    break
else:
  print("Not found")
```
